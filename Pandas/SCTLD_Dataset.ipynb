Looking at the info we first need to clean and fix certain things. There exist some null values in the MPA and MPA Year colummn wich will need to be taken care of. There also exist many type of categories that may be useful to seperate like for example: in group species there are "Dead Corals" which are corals that have completely died, there to is a pre and post outbreak that can be seperated as well.

Lets try and compare all the different MPA and how many dead corals they had

```python
import matplotlib.pyplot as plt

plt.bar(dead_coral['MPA'], dead_coral['Dead'])
plt.xticks(rotation=90)
plt.xlabel('Marine Protected Area')
plt.ylabel('Number of Dead Corals')
plt.title('Number of Dead Corals per MPA')
plt.show()
```

Lets try and compare helathy corals pre outbreak and post outbreak

```python
import numpy as np

# Get unique MPA values from both datasets
all_mpas = np.union1d(pre_outbreak['MPA'].unique(), post_outbreak['MPA'].unique())

# Create dictionaries to store healthy coral counts for each MPA
pre_outbreak_healthy = pre_outbreak.groupby('MPA')['Healthy'].sum().to_dict()
post_outbreak_healthy = post_outbreak.groupby('MPA')['Healthy'].sum().to_dict()

# Initialize lists to store data for plotting
mpas = []
pre_healthy_counts = []
post_healthy_counts = []

# Populate data lists, ensuring both datasets have the same MPAs
for mpa in all_mpas:
    mpas.append(mpa)
    pre_healthy_counts.append(pre_outbreak_healthy.get(mpa, 0))  # Use 0 if MPA not in pre_outbreak
    post_healthy_counts.append(post_outbreak_healthy.get(mpa, 0))  # Use 0 if MPA not in post_outbreak

# Plot the bars
plt.bar(mpas, pre_healthy_counts, alpha=1.0, label='Pre-Outbreak')
plt.bar(mpas, post_healthy_counts, bottom=pre_healthy_counts, alpha=0.5, label='Post-Outbreak')

plt.xticks(rotation=90)
plt.xlabel('Marine Protected Area')
plt.ylabel('Number of Healthy Corals')
plt.title('Number of Healthy Corals per MPA')
plt.legend()  # Add a legend to differentiate the bars
plt.show()
```
